```c++
TH2F *hist = new TH2F("hist", "Histogram", 100, -1, 1, 100, -1, 1);  //(Name, title, binsX, Xmin, Xmax, binsY, Ymin, Ymax)
TCanvas *c1 = new TCanvas();

gStyle->SetPalette(kRainBow); //this changes the palette of the graph to be rainbow colors

TRandom *randomNums = new TRandom(10);

for(int i = 0; i < 1e7; i++){
    
    double x = randomNums->Gaus();
    double y = randomNums->Gaus();
    
    hist->Fill(x, y);
}

//c1->SetLogz(); //sets the z axis to logarithm 
hist->SetContour(1000); //The scale on the right becomes smooth
hist->SetStats(0);
hist->GetXaxis()->SetTitle("x [cm]");
hist->GetYaxis()->SetTitle("y [cm]");
//hist->GetZaxis()->SetTitle("entries");

hist->Draw("colz"); //colz is used to get the color plot
c1->Draw();
```
