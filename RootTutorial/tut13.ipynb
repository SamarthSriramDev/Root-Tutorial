```c++
TCanvas *c1 = new TCanvas();

TF1 *func = new TF1("func","exp(-[1]*x)*[0]*sin(x)",0,10*TMath::Pi()); //(name,function,start,end) 
//TMath::Pi is used for using pi

func->SetParameter(0,1);
func->SetParameter(1,0.1);

func->SetTitle("");
c1->SetGridx();
c1->SetGridy();
func->GetXaxis()->SetTitle("Time");
func->GetYaxis()->SetTitle("Displacement");

func->Draw();

//c1->Print("func.pdf"); //this is used to save the graph as a pdf

c1->Draw();
```
