```c++
double x[5] = {1,2,3,4,5};
double y[5] = {1,4,9,16,25};

TGraph *gr = new TGraph(5,x,y); //The constructor takes the following args : no of points, x array, yarray

gr->SetMarkerStyle(2);
gr->SetMarkerSize(2);
gr->SetLineWidth(1);
gr->GetXaxis()->SetTitle("X Axis");
gr->GetYaxis()->SetTitle("Y Axis");
TCanvas *c1 = new TCanvas();
gr->Draw("AC*"); //We need to put A for all points, L for it to be a straight line between 
//data points and * as markers for points
//We can also use P to show no markers and C for a smooth curve between data points
c1->Draw();
```
